## Database Design

In this section, we learn how data is stored across **multiple tables**
in a database and how rows are **linked together**.

This linking of rows between tables is done by establishing
**relationships** in the database.

Understanding these relationships helps us:

- Organize data more cleanly
- Avoid duplication
- Design databases that are easy to understand and maintain
- Prepare for more advanced queries that span multiple tables


## Building a Data Model

Building a data model is the process of designing how data will be
organized, stored, and related inside a database.

---

## Step 1: Visualize the Data

The first step in building a data model is:

- Drawing a **picture** of the data objects used in the application
- Identifying what real-world objects the application needs to store
- Understanding how these objects are related to each other

This visual representation helps clarify the structure before writing any SQL.

---

## Representing Objects and Relationships

After identifying the data objects, we determine:

- How each object should be represented as a **table**
- What attributes (columns) each object needs
- How tables are connected using **relationships**

This ensures the database reflects the real-world structure of the data.

---

## Basic Rule of Data Modeling

> **Do not store the same string data more than once.  
> Use relationships instead.**

This means:
- Avoid duplicating the same information in multiple tables
- Store the data once and reference it using keys

Example:
- Store a user’s name once in a `Users` table
- Reference that user in other tables using a user ID

---

## One Real-World Thing = One Database Record

A core principle of good database design is:

> When there is one thing in the **real world**,  
> there should be **one copy** of that thing in the database.

Benefits of this rule:
- Prevents inconsistencies
- Makes updates easier
- Improves data integrity
- Reduces storage waste

---

## Why This Matters

Following these rules helps you:

- Design clean and understandable databases
- Avoid common design mistakes
- Build systems that scale well over time
- Write simpler and more powerful SQL queries

Good data modeling is the foundation of
reliable and maintainable database applications.


## Designing a Relational Data Model (Objects, Attributes, and Relationships)

Good database design starts by carefully examining **each piece of information**
and deciding how it should be stored. The goal is to create a clean, normalized,
and easy-to-maintain relational database.

---

## Object or Attribute?

For every piece of data (column), we must ask a key question:

> **Is this an object on its own, or just an attribute of another object?**

### Music Application Example

From a music dataset, we can identify the following **objects**:

- **Artist**
- **Album**
- **Track**
- **Genre**

These represent real-world entities and therefore should become **tables**.

On the other hand, some values are **attributes**, not standalone objects:

- **Len (length)** → attribute of a Track  
- **Rating** → attribute of a Track (or possibly a User–Track relationship)  
- **Count** → attribute (e.g., play count)

**Rule of thumb:**

- Objects → Tables  
- Attributes → Columns  

---

## Identifying Objects First

Before writing any SQL, we should:

1. Identify real-world objects
2. Decide which objects deserve their own tables
3. Avoid mixing multiple concepts into a single table

This makes the database schema easier to understand, extend, and debug.

---

## Defining Relationships Between Objects

Once the objects are clear, the next step is defining how they relate:

- A **Track belongs to an Album**
- An **Album belongs to an Artist**
- A **Track belongs to a Genre**

These are **belongs-to relationships** and are implemented using **foreign keys**,
not repeated text values.

---

## Why We Avoid Repeated Strings

Instead of storing text like:

- `"AC/DC"`
- `"Who Made Who"`
- `"Rock"`

in every row, we:

- Store each Artist once
- Store each Album once
- Store each Genre once
- Reference them using IDs

This follows a fundamental database design rule:

> **When there is one thing in the real world,
> there should be one copy of that thing in the database.**

---

## Track Attributes

A **Track** has descriptive attributes that belong directly to it:

- **Len** → track duration  
- **Rating** → user rating  
- **Count** → number of plays  

These values should be stored as columns in the `Track` table,
not as separate tables.

---

## Example Relational Schema

A simplified relational design might look like this:

- **Artist(id, name)**
- **Album(id, title, artist_id)**
- **Genre(id, name)**
- **Track(id, title, len, rating, count, album_id, genre_id)**

Each table has a clear responsibility and no duplicated data.

---

## Benefits of This Design

- No duplicated text data
- Easier updates (change once, reflected everywhere)
- Smaller database size
- Faster queries
- Clear and understandable schema
- Follows normalization principles

---

## Why This Matters

If everything is treated as a simple column:

- Data duplication increases
- Updates become error-prone
- The database becomes hard to scale

If objects and attributes are modeled correctly:

- Data is stored once
- Relationships are explicit
- Queries are more powerful and flexible

---

## Key Concepts: Primary Key, Logical Key, and Foreign Key

In relational data models, each table has specific types of keys:

- **Primary Key (PK)**:  
  Uniquely identifies each row.  
  Example: `Album.id`, `Track.id`.

- **Logical Key (Natural Key)**:  
  Represents meaningful real-world data.  
  Example: `Album.title`.

- **Foreign Key (FK)**:  
  Creates relationships between tables.  
  Example: `Track.album_id → Album.id`.

---

## Key Takeaway

> Good database design comes from asking the right questions about
> each piece of information before storing it.

Think first.  
Design second.  
Write SQL last.  
Optimize later.


## Implementing a Data Model in Tables

After designing a data model conceptually (objects and relationships),
the next step is to **implement that model using database tables**.

This means converting:
- Objects → Tables
- Attributes → Columns
- Relationships → Foreign Keys

---

## The “Belongs-To” Relationship

In our music data model:

- A **Track belongs to an Album**
- One Album can have **many Tracks**

This is a classic **one-to-many relationship**.

The rule is simple:

> The table on the “many” side stores the **foreign key**.

So:
- `Track` → many
- `Album` → one  
➡️ `Track` contains `album_id`

---

## Album Table

The `Album` table represents the Album object.

### Columns

  - A unique identifier for each album
- **title**

  - Human-readable album name

## Keys Explained

### Primary Key
- Uniquely identifies each row
- Usually an integer
- Example:
  - `Album.id`
  - `Track.id`

### Logical (Natural) Key
- Has real-world meaning
- Often text
- Example:
  - `Album.title`

### Foreign Key
- Creates a relationship between tables
- Stores the primary key of another table
- Example:
  - `Track.album_id → Album.id`

---

## Why We Use IDs Instead of Text

Instead of storing album titles repeatedly in the `Track` table:

❌ `"Who Made Who"`  
❌ `"Who Made Who"`  
❌ `"Who Made Who"`

and reference it using `album_id`.

This avoids:
- Data duplication
- Update anomalies
- Inconsistent values

## Relationship Summary

- **Album**
  - One album
  - Many tracks

- **Track**
  - Belongs to one album
  - References album using `album_id`

This structure enforces consistency and follows

## Key Takeaway

> Relationships in a data model become **foreign keys** in tables.

Design the relationship first.  
Implement it with keys second.

## Creating the Tables in SQLite

Now that we have defined the data model, we can create the tables in SQLite. Below is the SQL code for creating the `Artist`, `Album`, `Track`, and `Genre` tables:

```sql
-- Create the Artist table
CREATE TABLE Artist (
    id INTEGER PRIMARY KEY,   -- Primary Key
    name TEXT NOT NULL         -- Logical Key
);

-- Create the Genre table
CREATE TABLE Genre (
    id INTEGER PRIMARY KEY,   -- Primary Key
    name TEXT NOT NULL         -- Logical Key
);

-- Create the Album table
CREATE TABLE Album (
    id INTEGER PRIMARY KEY,        -- Primary Key
    title TEXT NOT NULL,            -- Logical Key
    artist_id INTEGER,              -- Foreign Key
    FOREIGN KEY (artist_id) REFERENCES Artist(id)  -- Relationship with Artist
);

-- Create the Track table
CREATE TABLE Track (
    id INTEGER PRIMARY KEY,        -- Primary Key
    title TEXT NOT NULL,            -- Logical Key
    rating INTEGER,                 -- Rating of the Track
    len INTEGER,                    -- Length of the Track (in seconds)
    count INTEGER,                  -- Play count of the Track
    album_id INTEGER,               -- Foreign Key
    genre_id INTEGER,               -- Foreign Key
    FOREIGN KEY (album_id) REFERENCES Album(id),   -- Relationship with Album
    FOREIGN KEY (genre_id) REFERENCES Genre(id)    -- Relationship with Genre
);


## Inserting Relational Data

Once the tables and relationships are created, the next step is to **insert data
in the correct order** so that foreign key relationships remain valid.

Because tables are connected using foreign keys, **parent tables must be filled
before child tables**.

The correct order is:

1. Artist  
2. Genre  
3. Album  
4. Track  

---

## Inserting Artists

Artists are independent objects and do not depend on any other table.

```sql
INSERT INTO Artist (name) VALUES ('Led Zeppelin');
INSERT INTO Artist (name) VALUES ('AC/DC');
```

Each artist is stored **once** and assigned a unique `id` automatically.

---

## Inserting Genres

Genres are also **standalone objects**, meaning each genre is unique and only stored once.

```sql
INSERT INTO Genre (name) VALUES ('Rock');
INSERT INTO Genre (name) VALUES ('Metal');
```
Again, each genre is stored once and referenced later by its `id`

---


## Inserting Tracks (Using Multiple Foreign Keys)

Tracks **belong to both an Album and a Genre**, so we must provide:

- `album_id` to link the track to an album.
- `genre_id` to link the track to a genre.

```sql
INSERT INTO Track (title, rating, len, count, album_id, genre_id)
VALUES ('Black Dog', 5, 297, 0, 2, 1);

INSERT INTO Track (title, rating, len, count, album_id, genre_id)
VALUES ('Stairway', 5, 482, 0, 2, 1);

INSERT INTO Track (title, rating, len, count, album_id, genre_id)
VALUES ('About to Rock', 5, 313, 0, 1, 2);

INSERT INTO Track (title, rating, len, count, album_id, genre_id)
VALUES ('Who Made Who', 5, 207, 0, 1, 2);
```


## Explanation

- `album_id` → references `Album(id)`
- `genre_id` → references `Genre(id)`

No album names or genre names are duplicated in the `Track` table.

Only numeric IDs are stored in the `Track` table, which references other tables through foreign keys.

---

## Why This Works

Instead of repeating text like:

- `"AC/DC"`
- `"Who Made Who"`
- `"Rock"`

We store them **once** in their respective tables and reference them using IDs.

This ensures:

- No duplicated data
- Consistent relationships
- Smaller and faster databases
- Easier updates (change once, reflected everywhere)

---

## Relationship Recap

- **Artist**
  - One artist → many albums

- **Album**
  - Belongs to one artist
  - One album → many tracks

- **Track**
  - Belongs to one album
  - Belongs to one genre

Each relationship is enforced using **foreign keys**.

---

## Key Takeaway

> When inserting relational data, always insert **parent tables** first,  
> then **child tables** using **foreign keys**.

- Design relationships first.  
- Insert data in the correct order.  
- Query across tables with confidence.



## Relational Power

One of the biggest advantages of relational databases is their **relational power**.

### Removing Replicated Data

By removing duplicated (replicated) data and replacing it with **references** to a single copy of each piece of information, we build a **web of information**.

This web allows the relational database to:

- Navigate relationships very quickly  
- Efficiently retrieve related data  
- Scale well, even with very large datasets  

Instead of copying the same values many times, we store them once and connect them using keys.

---

### Foreign Keys Create the Web

Often, when you want a piece of information, it does **not** come from just one table.

Instead, it comes from:

- Multiple tables
- Linked together using **foreign keys**

Foreign keys allow the database to follow relationships across tables and assemble the data you need.

---

### Why This Is Powerful

- Data is stored only once
- Relationships are explicit and enforced
- Queries can combine data from many tables
- Performance remains strong even as data grows

This interconnected structure is what gives relational databases their true power.

## The JOIN Operation

The **JOIN** operation is one of the most important features of relational databases.

### What JOIN Does

- A JOIN **links data across multiple tables**
- It is used as part of a `SELECT` query
- It allows you to combine related data that is stored in different tables

Instead of storing everything in one large table, we store data in separate tables and use JOINs to bring the data back together when needed.

---

### How JOIN Works

To use a JOIN correctly, you must tell the database:

- **Which tables are related**
- **Which keys connect them**

This is done using the **ON clause**.

Example idea (conceptual):

```sql
SELECT ...
FROM Track
JOIN Album ON Track.album_id = Album.id
```

The **ON** clause explains **how to use the keys** to match rows between tables.
It defines the condition that connects related records using primary and foreign keys.

---

## Why JOIN Is Powerful

- Keeps data **normalized** (no duplication)
- Allows **flexible and powerful queries**
- Lets you retrieve **meaningful real-world data**
  (e.g., track title + album name + artist name)

---

## Key Takeaway

> **JOIN** is how relational databases reconnect normalized data.

- **Foreign keys** define the relationships  
- **JOIN** uses those keys to combine data across tables



## JOIN Example: Linking Album and Artist

This example demonstrates how the **JOIN** operation reconnects normalized data
stored across multiple tables using **foreign keys**.

---

## The Tables Involved

### Artist Table

| id | name         |
|----|--------------|
| 1  | Led Zeppelin |
| 2  | AC/DC        |

### Album Table

| id | artist_id | title         |
|----|-----------|---------------|
| 1  | 2         | Who Made Who  |
| 2  | 1         | IV            |

- `Artist.id` is the **primary key**
- `Album.artist_id` is a **foreign key** that references `Artist.id`

## What We Want to See

We want a result that shows:

- Album title
- Artist name

In a **single result set**, even though the data lives in two different tables.

---

## How the Tables Are Linked

- `Album.artist_id` → points to → `Artist.id`
- This relationship tells us **which artist created each album**

## The JOIN Query

```sql
SELECT Album.title, Artist.name
FROM Album
JOIN Artist
ON Album.artist_id = Artist.id;
```

This query returns **each track’s title together with its genre name**
by joining the `Track` and `Genre` tables.

## What This Query Answers

> **“What is the genre of each track?”**

## JOIN Without an ON Clause (Cartesian Product)

Consider the following query:

```sql
SELECT Track.title,
       Track.genre_id,
       Genre.id,
       Genre.name
FROM Track
JOIN Genre;
```
## What This Query Does

This query joins the `Track` and `Genre` tables **without an `ON` clause**.

When a `JOIN` is written without specifying how the tables are related,  
the database produces a **Cartesian product**.

---

## What Is a Cartesian Product?

A Cartesian product means:

> **Every row in `Track` is combined with every row in `Genre`.**

If:

- `Track` has **4 rows**
- `Genre` has **2 rows**

The result will contain:

4 × 2 = 8 rows

## Why This Is Usually Wrong

Without an `ON` clause:

- The database does not know **how rows should be matched**
- **Foreign keys are ignored**
- **Invalid combinations** of rows are produced
- The result does **not represent real-world relationships**

This leads to misleading and incorrect query results.

---

## The Correct Way: Use an ON Clause

To correctly combine data from related tables, you must explicitly
define how the tables are linked using an `ON` clause.

```sql
SELECT Track.title, Genre.name
FROM Track
JOIN Genre
ON Track.genre_id = Genre.id;
```

This tells the database:

- Use **Track.genre_id** (foreign key)
- Match it to **Genre.id** (primary key)
- Return **only meaningful relationships**

## Key Takeaway

- **Foreign keys define relationships**
- **JOIN uses those keys to recombine normalized data**
- **JOIN without ON** → Cartesian product (all possible combinations)
- **JOIN with ON** → Correct relational data

Always define **how tables are related** when using `JOIN`.

Normalize your tables.  
Use `JOIN` to retrieve meaningful results.
